```python
import chess
from chesstransformer.models.tokenizer.position_tokenizer import PostionTokenizer
```

```python
# create a chess board and set it to a specific position
board = chess.Board()

starting_position = "e2e4 e7e5 g1f3 b8c6 f1c4 g8f6"

for move in starting_position.split():
    board.push_uci(move)

print(board)
```

## Board encoding example

```python
encoder = PostionTokenizer()
encoded = encoder.encode(board)
print(f"Encoded: {encoded}")
```

## Board decoding example

```python
decoded = encoder.decode(encoded)
print(f"Decoded:\n{decoded}")
```
